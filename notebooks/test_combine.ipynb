```python
def extract_cycle(s):
    if "cycle" in s.lower():
        left, right = s.split(":", 1)
        return left.strip(), right.strip()
    return None, s.strip()
    
def extract_author(s):
    if ", by" in s:
        left, right = s.split(", by", 1)
        return left.strip(), right.strip()
    return s.strip(), None

def get_short_title(s):
    if '(' in s:
        s=re.sub(r'\s*\(.*?\)', '', s).strip()
        # re.sub(pattern, replacement, string)
        # 0+ space, (, 1+ letters, )
    if "," in s:
        first, rest=s.split(',',1)
        return first

    return s

# s="Cycle of Arthur: Iwein, by Hartmann von Aue"
# s="Life of Harold, king of England"
s="Ricciardetto, by Niccolo Fortiguerra"

cycle, title_author = extract_cycle(s)
title, author =extract_author(title_author)
short_title=get_short_title(title)
print("cycle:",cycle)
print("title:",title)
print('short title:', short_title)
print("author:",author)
```

```python

from rapidfuzz import process, fuzz

def get_mss_info(title_theme, index_table_clean, score_cutoff=70):

    # 1. EXACT match
    if title_theme in index_table_clean:
        title_index=title_theme
        return title_index, index_table_clean[title_theme]
    
    # 2. FUZZY match
    result = process.extractOne(
        title_theme,
        index_table_clean.keys(),
        scorer=fuzz.ratio,
        score_cutoff=score_cutoff
    )
    
    if result:
        best_match, score, _ = result
        title_index=best_match
        # print(f"Fuzzy match: {best_match} (score={score})")
        return title_index, index_table_clean[best_match]
    # print(f"[No match] found for {title}")
    return None, []


title="Roman de la Bose"#"Iliaca"#"Heroica"
title, mss_info=get_mss_info(title, index_table_clean)
print(title,len(mss_info))
for info in mss_info:
    print(info)
```
